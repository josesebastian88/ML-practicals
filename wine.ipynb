# load the wine dataset
# using the PCA estimator reduce the data into 1 dimension
# show the components and explained variance ratio
# print the shape of the reduced data

```python
import pandas as pd
wine = pd.read_csv("wine.csv")
wine
```

```python
wine.head()
```

```python
features = wine.drop("Wine",axis=1)
features.head()
```

```python
#shape of original data
features.shape
```

```python
#reduction of dimensions
import seaborn as sns
from sklearn.decomposition import PCA
model = PCA(n_components=1)
model.fit(features)
```

```python
features_1D = model.transform(features)
```

```python
features_1D
```

```python
#shape of reduced data
features_1D.shape
```

```python
#print eigen vectors
print(model.components_)
```

```python
#print explained variance ratio
print(model.explained_variance_ratio_)
```
